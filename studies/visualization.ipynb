```python
import pandas as pd
from sklearn.manifold import TSNE

from bokeh.io import show, output_notebook, push_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
output_notebook()
```

```python
genres_count = pd.read_pickle('../temp/genres_count.pickle')
synopsis_count = pd.read_pickle('../temp/synopsis_count.pickle')
```

```python
def plot_graph(df):
    model = TSNE(n_components=2, learning_rate=200, random_state=0)
    tsne_features = model.fit_transform(df.drop(['id', 'title'], axis=1))

    df['X'] = tsne_features[:,0]
    df['Y'] = tsne_features[:,1]
    
    source = ColumnDataSource(df)
    plot = figure(x_axis_label = "T-SNE 1", y_axis_label = "T-SNE 2", width = 900, height = 400)
    plot.circle(x = "X", y = "Y", source = source, size = 10, color = '#FF7373', alpha = .8)

    hover = HoverTool(tooltips = [('Title', '@title'), ('ID', '@id')])
    plot.add_tools(hover)
    
    show(plot)
```

```python
plot_graph(genres_count.copy())
```

```python
plot_graph(synopsis_count)
```
